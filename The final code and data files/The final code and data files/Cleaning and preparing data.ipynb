```python
import pickle
import pandas as pd
from faker import Faker
import random
```

```python
#open pkl file
with open('movies.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
pd.set_option('display.max_colwidth', None) # Display full contents of cells
```

```python
#converting the data to datafram
df = pd.DataFrame(data)
```

```python
# copying the dataset
df_movies_clean = df.copy()

# deleting the column
df_movies_clean.drop(['category'], axis=1, inplace=True)

# heading
df_movies_clean.head()
```

```python
#cleaning
#deleting the square bracets 
df_movies_clean['tags']=df_movies_clean['tags'].astype(str).str.replace('[', '', regex = True)
df_movies_clean['tags']=df_movies_clean['tags'].astype(str).str.replace(']', '', regex = True)
```

```python
#slove a problem in some movies name 
#some movies has in its name (,) comma which make confusion when adding multiple movies together in one list (it will consider the movie name as two movies)
#for the above reason we deleted all the commas from the movies names
df_movies_clean['title']=df_movies_clean['title'].astype(str).str.replace(",", '', regex = True)
```

```python
#saving the data as CSV
df_movies_clean.to_csv(r'movies_data.csv',index=False)
```

```python
#opening the csv file
df_movies= pd.read_csv('movies_data.csv', sep=',')
#df_movies.drop(['Unnamed: 0'], axis=1, inplace=True)  #in case we didn't use index=False when saveing as csv
df_movies.head()

#test
df_movies['description'][df_movies['title']=='Lady Chatterley\'s Lover']
```

```python
# determine the most used tags values counts to use them in the interest for the fake users
df_tags= df_movies.copy()
df_tags['tags'] = df_tags['tags'].str.split(', ')
df = df_tags.explode('tags')
tag_counts = df['tags'].value_counts()
tag_counts
```

```python
#generate fake users
fake = Faker() 
fake.seed_instance(0)
random.seed(0)

users = []

# Loop through user IDs from 0 to 99
for user_id in range(10000,10100):
    
    password=fake.password()

    # Generate user data
    age = fake.random_int(min=18, max=70)
    
    gender = fake.random_element(elements=("Male", "Female"))
    
    first_name=fake.first_name_male() if gender=="Male" else fake.first_name_female()
    
    interests = []
    for i in range(3):
        interest = fake.random_element(elements=("Action", "Comedy", "Drama", "Thrilling",'Mysterious', \
                                                 'Psychological','Romantic','Local','Epic','Kids'))
        if interest not in interests:                                        
            interests.append(interest)
        
    view_history = []
    for j in range(5):
        movie = random.choice(df_movies['title'])
        # add the title to the view_history if it is not already in the list
        if movie not in view_history:
            view_history.append(movie)
            
    email=fake.email()
    
    mother_language=fake.random_element(elements=("English","Māori","French", "Arabic", "Russian", "Chinese", "other"))
    
    
    # Create a new user dictionary and add the key-value pairs
    users_data = {}
    users_data["user_id"] = user_id
    users_data["first_name"]= first_name
    users_data["password"]= password
    users_data["age"] = age
    users_data["gender"] = gender
    users_data["interests"] = interests
    users_data["view_history"] = view_history
    users_data["email"]= email
    users_data["mother_language"]= mother_language
    
    # Append the user data dictionary to the users list
    users.append(users_data)
    
    
df_users = pd.DataFrame(users)
df_users.head()
```

```python
#clean the data
#deleting the square bracets 
df_users['interests']=df_users['interests'].astype(str).str.replace('[', '', regex = True).str.replace(']', '', regex = True)
df_users['view_history']=df_users['view_history'].astype(str).str.replace('[', '', regex = True).str.replace(']', '', regex = True)
```

```python
#deleting the ('') Quotation Mark from the values in the interest column, but we didn't do this for the values in the view_history column because
#some movies have (') Apostrophe in its name and we don't want to delete it, so we will clean this when preparing the data to the recommendation algorithem 
df_users['interests']=df_users['interests'].astype(str).str.replace("'", '', regex = True)
```

```python
#the authentication function, that will check if the user's Id, name and password are valid
#we need it for the interface to login user 

def authenticate(user_id,username, password):
    
    #check the user_id
    ids= df_users_data['user_id'].tolist()
    #check that the name is related to the user_id
    names=df_users_data['first_name'][df_users_data['user_id']==user_id].iloc[0]
    #check that the password is related to the user_id
    passwords=df_users_data['password'][df_users_data['user_id']==user_id].iloc[0]
    
    if  user_id in ids and username == names  and password == passwords:
        #print('correct')
        return True
    else:
        #print('wrong')
        return False
    
    
```
